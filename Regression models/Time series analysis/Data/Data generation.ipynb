## 1. May (22 to 31)

```python
df = pd.read_csv('/content/drive/MyDrive/Biogas RMS project/Datasets/Lab dataset/may_all_data.csv')
df.head()
```

### Grouping 1440 (60 minutes * 24 hours) values and taking average

```python
may = df.groupby(np.arange(len(df))//1440).mean()
may
```

## 2. June (1 to 30)

```python
df = pd.read_csv('/content/drive/MyDrive/Biogas RMS project/Datasets/Lab dataset/june_all_data.csv', na_values='NA')
df.head()
```

## Combining all the months' data

```python
combined_df = may
combined_df = combined_df.append(june, ignore_index=True)
combined_df = combined_df.append(july, ignore_index=True)
combined_df = combined_df.append(august, ignore_index=True)
combined_df = combined_df.append(sept, ignore_index=True)
combined_df.head()
```

### Extracting into individual reactor's data

```python
reactor1_data = combined_df.iloc[:][["PH100", "PI100", "TI100_SP", "TI100_PV"]]
reactor2_data = combined_df.iloc[:][["PH200", "PI200", "TI200_SP", "TI200_PV"]]
reactor3_data = combined_df.iloc[:][["PH300", "PI300", "TI300_SP", "TI300_PV"]]
reactor4_data = combined_df.iloc[:][["PH400", "PI400", "TI400_SP", "TI400_PV"]]
```

```python
# convert to CSV
reactor1_data.to_csv('reactor1_data.csv')
reactor2_data.to_csv('reactor2_data.csv')
reactor3_data.to_csv('reactor3_data.csv')
reactor4_data.to_csv('reactor4_data.csv')
```
